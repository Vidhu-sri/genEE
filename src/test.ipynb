```python


import os, re
from dotenv import load_dotenv


load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")

class LLM:
    def __init__(self, model="qwen/qwen3-32b", temperature=0.7, max_tokens=512):
        self.llm = ChatGroq(model=model, groq_api_key = groq_api_key,temperature=temperature, max_tokens=max_tokens)

    def generate(self, prompt: str) -> str:
        resp = self.llm.invoke(prompt)
        return resp.content.strip()

    def relevance_score_1to10(self, prompt: str) -> float:
        # deterministic scoring
        scorer = self.llm.bind(temperature=0.0, max_tokens=8)
        resp = scorer.invoke(
            "Reply with a single integer 1..10. No other text.\n\n" + prompt
        )
        txt = resp.content.strip()
        m = re.search(r"\b(10|[1-9])\b", txt)
        return float(m.group(1)) if m else 5.0

    def generate_list(self, prompt: str, k: int = 10):
        resp = self.llm.invoke(prompt)
        lines = [ln.strip("-• ").strip() for ln in resp.content.splitlines() if ln.strip()]
        return lines[:k]
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ctr_theoretical(score, T, RS=11, K=3):
    # equal scores for all K shown questions
    num = np.exp(score / T)
    den = K * np.exp(score / T) + np.exp(RS / T)
    return 100.0 * (num / den)

def main():
    RS = 11
    K = 3
    Ts = np.logspace(-1, 1, 200)  # 0.1 to 10
    plt.figure()
    for s in range(1, 11):
        ctrs = [ctr_theoretical(s, T, RS=RS, K=K) for T in Ts]
        plt.plot(Ts, ctrs, label=f"score = {s}")
    plt.xscale("log")
    plt.xlabel("Softmax Temperature (T)")
    plt.ylabel("CTR (%)")
    plt.axvline(1.5, linestyle="--")  # paper uses T=1.5
    plt.title(f"Theoretical CTR values with RS={RS}, K={K}")
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

def load_and_aggregate(csv_path: Path):
    df = pd.read_csv(csv_path)

    # average across topics for each iteration
    g = (
        df.groupby("iteration", as_index=False)
          .agg(avg_ctr_pct=("avg_ctr_pct", "mean"))
    )
    return g

def plot_compare(csv1, label1, csv2, label2, title=None):
    g1 = load_and_aggregate(Path(csv1))
    g2 = load_and_aggregate(Path(csv2))

    plt.figure(figsize=(7, 4))

    plt.plot(
        g1["iteration"], g1["avg_ctr_pct"],
        marker="o", linewidth=2, label=label1
    )
    plt.plot(
        g2["iteration"], g2["avg_ctr_pct"],
        marker="s", linewidth=2, label=label2
    )

    plt.xlabel("Iteration")
    plt.ylabel("Average CTR (%)")
    plt.grid(True, alpha=0.3)
    plt.legend()
    if title:
        plt.title(title)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    plot_compare(
        csv1=r"C:\Users\Lenovo\Desktop\research\genEE\results\ctr_results_openai.csv",
        label1="Model A (GPT-Turbo)",
        csv2=r"C:\Users\Lenovo\Desktop\research\genEE\results-Qwen32B-instruct.csv",
        label2="Model B (Qwen32B-Instruct)",
        title="Average CTR comparison"
    )
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
from pathlib import Path

def plot_compare(openai_csv_path, qwen_csv_path):
    openai_df = pd.read_csv(openai_csv_path)
    qwen_df = pd.read_csv(qwen_csv_path)

    # find common (topic, domain) pairs
    common = set(
        zip(openai_df.topic, openai_df.domain)
    ).intersection(
        set(zip(qwen_df.topic, qwen_df.domain))
    )

    for topic, domain in sorted(common):
        oa = openai_df[(openai_df.topic == topic) & (openai_df.domain == domain)]
        qw = qwen_df[(qwen_df.topic == topic) & (qwen_df.domain == domain)]

        plt.figure()
        plt.plot(oa.iteration, oa.avg_ctr, label="openai")
        plt.plot(qw.iteration, qw.avg_ctr, label="qwen")
        plt.title(f"{topic} ({domain})")
        plt.xlabel("iteration")
        plt.ylabel("avg_ctr")
        plt.legend()
        plt.show()


p1 = r"C:\Users\Lenovo\Desktop\research\genEE\results\ctr_results_openai.csv"
p2 = r"C:\Users\Lenovo\Desktop\research\genEE\results-Qwen32B-instruct.csv"


plot_compare(Path(p1), Path(p2))
```
